# Taking a unique product to further analysis for y_train_T2_single_product (input required)

# Input for one product (not channels)

```python
#Create list of products
y_train_T2_product_list = pd.unique(y_train_T2['key'])

#First product
#YOUR INPUT IS NEEDED: select index of the product that will be analysed further. 
#The final forecast for this product will be saved in a file y_test_T2. 
y_train_T2_single_product_number = y_train_T2_product_list[0]
print(y_train_T2_single_product_number)


#original set for first product
# y_train_T2_single_product = y_train_T2.loc[y_train_T2['key'] == y_train_T2_single_product_number, 'y'].dropna()
y_train_T2_single_product = y_train_T2[y_train_T2['key'] == y_train_T2_single_product_number].dropna()
print(len(y_train_T2_single_product))

#Create train and test sets
x = int(0.8 * y_train_T2_single_product.shape[0])
y_train_T2_single_product_train = y_train_T2_single_product.iloc[:x]
y_train_T2_single_product_test = y_train_T2_single_product.iloc[x:]
y_train_T2_single_product.shape[0] == y_train_T2_single_product_train.shape[0] + y_train_T2_single_product_test.shape[0]
```

# Taking a unique product to further analysis for product from merged files: X_train_T2 and y_train_T2 files  (input required)

# Input for one product in merged file (with channels)

```python
#merging two files X_train_T2 and y_train_T2 to obtain multichannel inputs for forecasts. 
#This file will be used for multiple regressors

X_train_T2_y_train_T2_merged = pd.merge(left=X_train_T2, right=y_train_T2, on=['key', 'date'], how='inner')

#First product
#YOUR INPUT IS NEEDED: select index of the product that will be analysed further. 
#The final forecast for this product will be saved in a file y_test_T2. 
X_train_T2_y_train_T2_merged_product_number = y_train_T2_product_list[0]


X_train_T2_single_product_MERGED = X_train_T2_y_train_T2_merged[X_train_T2_y_train_T2_merged['key'] == X_train_T2_y_train_T2_merged_product_number]
print(X_train_T2_single_product_MERGED)

```

# Single product forecast

```python
#dataset for a selected ptoduct: y_train_T2_single_product_train
#Fit method allows to pass historical dataframe, first we need to describe 'ds' and 'y'
y_train_T2_single_product_train['ds'] = y_train_T2_single_product_train['date']

#model
m = Prophet()
m.fit(y_train_T2_single_product_train[['ds', 'y']])

#Create a future dataframe with monthly frequency for the next 12 months
future = m.make_future_dataframe(periods=12, freq='MS')

#Make a prediction for the future dataframe
forecast = m.predict(future)
# print(forecast)

fig = m.plot(forecast)

#adjust date in text file to beable to display in plot below. Date in forecast and test files mudt have the same format.
y_train_T2_single_product_test = dl.parse_dates(y_train_T2_single_product_test) #parsing

plt.gca()
ax = fig.axes

ax[0].plot(y_train_T2_single_product_test['date'], y_train_T2_single_product_test['y'], marker='.', color='red')

fig = m.plot_components(forecast)

#TODO:
#1. nie ruszam NaN, ruszam regressory i patrzę czy prophet radzi sobie z regresorami
#2. Jeśli prophet nie radzi sobie z NaN, wtedy wywalamy punktowo
#3. Dodawać po jednym regressorze (channelu) w pętli for
#pobawić się z add.regresors  -dodawać po jednym channelu i zobaczyli które channele mają sens, a które nie
#https://facebook.github.io/prophet/docs/seasonality,_holiday_effects,_and_regressors.html#coefficients-of-additional-regressors
```
